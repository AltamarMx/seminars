```python

comment_words = '''
Barrios, 
Guillermo Ramirez,
Mariana Lopez,
Jose Escobar,
Edgar Gallardo,
Uriel Rodriguez,
Giovanni Velazquez,
Julio Landa,
Diego Salgado,
Miguel Gaspar,
Fernando de Olarte,
Marco Morales,
Oliver Stif,
Abraham Gonzalez,
Rodriguez,
Victor Zamora,
Hector Cortes,
Gonzalo Gaona,
Debra Lopez,
Jorge Rojas,
Guadalupe Huelsz,
Ramon Tovar,
Kevin Alquicira,
Benjamin Santamaria,
UTEZ,
IER,
ITE,
ITZ,
UAG
'''
```

```python
#| label: fig-wordcloud
#| fig-cap: "Histrico de participantes"

wordcloud = WordCloud(width = 600, height = 600,
				background_color ='white',
				stopwords = STOPWORDS,
				min_font_size = 10).generate(comment_words)

# plot the WordCloud image					 
fig, ax = plt.subplots(figsize = (8, 8), facecolor = None)
ax.imshow(wordcloud)
ax.axis('off');
# fig.tight_layout(pad = 0)

```
