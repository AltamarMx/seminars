```python
from iertools.read import read_epw
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import calplot
import numpy as np
```

```python
plt.style.use('matplotlib.mplstyle')
```

```python
f = 'data/Temixco_2018CST.epw'
df = read_epw(f,alias=True)
```

```python
#| label: fig-temixco-to
#| fig-cap: Temperatura ambiente en Temixco, Morelos.
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(df.To,'k-',alpha=0.3)
ax.plot(df.To.resample('D').mean())
ax.set_ylim(4,40)
ax.set_ylabel('Temperatura [$^oC$]')
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(mdates.AutoDateLocator()))
```
