## Datos




```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
anio = '2023'

f = '../data/001_raw/001_CompletarFechas/eplusout.csv'
data = pd.read_csv(f)
data.rename(columns={'Date/Time':'time'},inplace=True)


def convierte(fechas):
    fecha, hora = fechas.split()
    mes_dia = fecha.split('/')
    if hora == '24:00:00':
        fecha_str = f"{anio}-{mes_dia[0]}-{mes_dia[1]} 00:00:00"
        return pd.Timestamp(fecha_str) + pd.Timedelta(days=1)
    else:
        fecha_str = f"{anio}-{mes_dia[0]}-{mes_dia[1]} {hora}"
        return pd.Timestamp(fecha_str)
    
anio = 2020
data['time'] = data['time'].apply(convierte)
data.set_index('time',inplace=True)
```

```python
data.plot(subplots=True,figsize=(14,8))
```

```python
data.to_csv('../data/002_processed/eplusout.csv')
```
