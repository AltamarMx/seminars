````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x   = np.linspace(-np.pi,np.pi, 100)
y1  = np.sin(x)
y2  = 100 * np.cos(x)
````

````python
#x,y1,y2
````

````python
fig, ax1 = plt.subplots(figsize=(10,4))

ax1.plot(x,y1,'r.',label = "Uno")
ax1.twinx()
ax2 = ax1.twinx()

ax2.plot(x,y2,'b+',label = "Cien")

g1, = ax1.plot(x,y1,'r.',label = "Uno")
g2, = ax2.plot(x,y2,'b+',label = "Cien")

ax1.set_ylabel("Uno")
ax2.set_ylabel("Cien")

ax1.legend(loc="upper left")
ax2.legend();
````

````python
lineas = [g1, g2]
lineas
````

````python
etiquetas = [linea.get_label() for linea in lineas]
etiquetas
````

````python
fig, ax1 = plt.subplots(figsize=(10,4))

g1, = ax1.plot(x,y1,'r.',label = "Uno")
ax2 = ax1.twinx()
g2, = ax2.plot(x,y2,'b+',label = "Cien")

ax1.set_ylabel("Uno")
ax2.set_ylabel("Cien")

lineas = [g1, g2]
etiquetas = [linea.get_label() for linea in lineas]
ax1.legend(lineas,etiquetas);
````
